```python
n = int(input("Введите количество спортсменов: "))
results = [ ]
result = 0
x = 0
for i in range(n):
 x=x+1
 print(("Введите результат спортсмена номер"),(x),(":"))
 result = int(input())
 results.append(result)
for i, result in enumerate(results, 1):
 print(f"Спортсмен {i}: {result} сек")
 best_result = min(results)
 best_result_index = results.index(best_result) + 1
print(f"Лучший результат у спортсмена {best_result_index}: {best_result} сек")
```
